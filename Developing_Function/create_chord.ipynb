```python
def change_pitch_scale(scale_name, pitch):
    delta = 0
    if scale_name == "C":
        delta = 0
    elif scale_name == "bC":
        delta = -1
    elif scale_name == "bD":
        delta = 1
    elif scale_name == "D":
        delta = 2
    elif scale_name == "bE":
        delta = 3
    elif scale_name == "E":
        delta = 4
    elif scale_name == "F":
        delta = 5
    elif scale_name == "bG":
        delta = 6
    elif scale_name == "G":
        delta = 7
    elif scale_name == "bA":
        delta = 8
    elif scale_name == "A":
        delta = 9
    elif scale_name == "bB":
        delta = 10
    elif scale_name == "B":
        delta = 11
    pitch = pitch + delta
    return pitch 
```

```python
def change_scale(pretty_midi, scale_name):
    for track in pretty_midi.instruments:
        for note in track.notes:
            note.pitch = change_pitch_scale(scale_name, note.pitch)
    return pretty_midi
```
