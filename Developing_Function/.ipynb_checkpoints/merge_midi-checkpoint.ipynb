```python
from mido import MidiFile

mid2 = MidiFile('MuseGanTest.mid')
mid = MidiFile('five_chord.mid')
```

```python
etime=0
for msg in mid.tracks[0]:
    if not msg.is_meta:
        if (msg.type == 'note_on' or msg.type == 'note_off'):
            etime += msg.time
```

```python
for track in mid2.tracks:
    for msg in track:
        if not msg.is_meta:
            if (msg.type == 'note_on' or msg.type == 'note_off'):
                msg.time=int(msg.time*2.5)
```

```python
for track in mid2.tracks:
    track[0].time=etime+480
    mid.tracks.append(track)
```

```python
mid.save('merge_test.mid')
```
