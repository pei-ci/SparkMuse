```python
from mido import MidiFile, tick2second
mid2 = MidiFile('nine_chord.mid')
mid = MidiFile('merge_test.mid')
```

```python
def get_etime(midi):
    max_etime=0
    for track in midi.tracks:
        etime=0
        for msg in track:
            etime+=msg.time
        if etime>max_etime:
            max_etime=etime
    return max_etime
```

```python
if len(mid2.tracks) > 1:
    for track in mid2.tracks:
        for msg in track:
            if not msg.is_meta:
                if (msg.type == 'note_on' or msg.type == 'note_off'):
                    msg.time = int(msg.time*2.2)
```

```python
mid_etime = get_etime(mid)
for track in mid2.tracks:
    track[0].time=mid_etime+480
    #mid2 append to mid
    mid.tracks.append(track)
```

```python
mid.save('merge_test2.mid')
```

```python
import mido
from mido import MidiFile

mid = MidiFile("Test.mid")

tempo = mido.bpm2tempo(131)

for i, track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i, track.name))
    passed_time = 0
    for msg in track:
        ab_time = mido.tick2second(msg.time, mid.ticks_per_beat, tempo)
        # real_time就是每一个事件在整个midi文件中的真实时间位置
        real_time = ab_time + passed_time
        passed_time += ab_time
        if msg.type == "note_on":
             print(msg, " read time=" + str(round(real_time, 3)))
```
