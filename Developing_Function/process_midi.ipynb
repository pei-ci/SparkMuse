```python
import mido

instrument_list=['Piano', 'Drums', 'Guitar', 'Bass']

def change_midi_velocity(read_file_path, save_file_path):
    # read
    mid = mido.MidiFile(read_file_path)
    for i, track in enumerate(mid.tracks):
        # 如果不是鋼琴不調整音量
        if track.name not in instrument_list:
            continue
        for msg in track:
            if not msg.is_meta:
                if (msg.type == 'note_on' and msg.velocity != 0):
                    msg.velocity = 64
                    #if msg.velocity > 127:
                     #   msg.velocity = 127
    # save
    mid.save(save_file_path)


# 代表主函數
if __name__ == '__main__':
    # 調用函數
    change_midi_velocity("SurpriseTest.mid", "SurpriseTest.mid")
```
